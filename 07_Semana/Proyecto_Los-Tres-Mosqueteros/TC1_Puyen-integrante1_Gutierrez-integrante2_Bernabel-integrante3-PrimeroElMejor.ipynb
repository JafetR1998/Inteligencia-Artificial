```python
#Diccionario con el nombre de las filas con sus valores computacionales
pos_fc = {"A" : 0, "B" : 1, "C" : 2, "D" : 3, "E" : 4}

#Diccionario con los valores invertidos de pos_fc
pos_cf = dict([(v, k) for k, v in pos_fc.items()])

#Diccionario con los movimientos posibles en el plano cartesiano
movimientos = {"arriba" : (-1, 0), "abajo" : (+1, 0), "izquierda" : (0, -1), "derecha" : (0, +1)}

#Lista generada a partir de las 4 keys del diccionario de movimientos
orientaciones = list(movimientos.keys())
```

```python
"""Crea el tablero inicial del juego, definiendo al mismo como global para que no sea pasado como
parámetro, a fin de reducir la cantidad de los mismos en las funciones, tal que, el tablero es una
lista de listas, con 'G' representando a los guardias; 'M', a los mosqueteros y posteriormente,
cuando el juego sea ejecutado, se tendrán a '-' como los espacios vacíos que se dejarán."""
def crear_tablero():
    global tablero
    tablero = [["G", "G", "G", "G", "M"],
            ["G", "G", "G", "G", "G"],
            ["G", "G", "M", "G", "G"],
            ["G", "G", "G", "G", "G"],
            ["M", "G", "G", "G", "G"]]

"""Reemplaza el tablero inicial con un nuevo tablero a través de un deepcopy"""
def set_tablero(nuevo_tablero):
    from copy import deepcopy
    global tablero
    tablero = deepcopy(nuevo_tablero)

"""Se obtiene a la versión vigente del tablero global"""
def get_tablero():
    return [[casillero for casillero in fila] for fila in tablero]

"""Se devuelve una lista con las posiciones vigentes de los mosqueteros en cierta jugada"""
def posiciones_mosquetero():
    posiciones_mosquetero = [i for i in lista_posiciones() if ver_casillero(i) == "M"]
    return posiciones_mosquetero

"""Se crea un contador con las direcciones que tendrá la pc, a fin de controlar el comportamiento 
de la estrategia seleccionada por su dificultad, ya sea, aleatoria o por Primero el mejor"""
def crear_contador():
    global contador_pc
    contador_pc = {"arriba" : 0, "abajo" : 0, "izquierda" : 0, "derecha" : 0}

"""A partir de una dirección indicada para un movimiento, se agregará al cómputo del contador de
jugadas que ha realizado la pc, con lo cual, se cumplirá el objetivo propuesto del comportamiento"""
def agregar_movimiento(direccion):
    if direccion == "arriba":
        contador_pc["arriba"] = contador_pc["arriba"] + 1
    elif direccion == "abajo":
        contador_pc["abajo"] = contador_pc["abajo"] + 1
    elif direccion == "izquierda":
        contador_pc["izquierda"] = contador_pc["izquierda"] + 1
    elif direccion == "derecha":
        contador_pc["derecha"] = contador_pc["derecha"] + 1
    return contador_pc

"""Se imprime el tablero, considerando al diccionario pos_cf para las iteraciones de las filas y
respectivas columnas que dispone el mismo de 5 x 5, el cual, ahora se podrá visualizar en pantalla"""
def imprimir_tablero():
    contador = 0
    print("    1  2  3  4  5")
    print("   --------------")
    for fila in tablero:
        print(pos_cf[contador] + " |", end = " ")
        for columna in fila:
            print(columna + " ", end = " ")
        contador = contador + 1
        print()

"""Verifica si, dada una posición (x, y) del tablero de 5 x 5 con entradas de números enteros, estén
en el rango de 0 a 4, tal que, si se cumple la condición, se retornará True; caso contrario, False"""
def es_pos_legal(pos):
    if pos[0] in range(0, 5) and pos[1] in range(0, 5):
        return True
    else:
        return False        
    
"""Dada un string de 2 caracteres (letra y número) como B4, se convertirá a su notación a una tupla
que indicará las coordenadas computacionales, ya sea, en el ejemplo (1,3), siempre y cuando, el par
ordenado resultante sea una posición legal; esto es, que esté entre los valores de 0 y 4"""
def str_a_pos(s):
    x = int(ord(s[0]) - 65)
    y = int(s[1]) - 1
    if es_pos_legal((x, y)) == False:
        raise ValueError("La posición es inválida")
    return((x, y))

"""Realiza el proceso contrario que la función str_a_pos, al ser que, devuelve, dada una posición,
un string comprendido por 2 caracteres; por ejemplo, tomando el caso previo, si se da la coordenada
(1, 3), se espera que se devuelva B4, ya que, ambos son enteros y representan una posición válida"""
def pos_a_str(pos):
    x = chr(pos[0] + 65)
    y = str(int(pos[1]) + 1)
    if es_pos_legal(pos) == False:
        raise ValueError("La posición es inválida")
    return("{0}{1}".format(x, y))

"""Se permite ver el contenido de un casillero dada una posición, a saber si en el mismo se encuentra
un mosquetero, un guardia o si el espacio está vacío, siempre y cuando, se determine una posición
correcta; caso contrario, se generará la excepción IndexError sobre el espacio del tablero vigente"""
def ver_casillero(pos):
    pos_f = pos[0]
    pos_c = pos[1]
    try:
        casillero = tablero[pos_f][pos_c]
        return casillero
    except IndexError as e:
        pass

"""Devuelve la lista de todas las posiciones del tablero, con la finalidad de que sean iteradas en
funciones posteriores, considerando así, 25 , sean, por ejemplo, (0, 0), (0, 1), (0, 2), ..., (4, 4)"""
def lista_posiciones():
    return [(i, j) for i in range(5) for j in range(5)]

"""Retorna las coordenadas de la posición adyacente, dado un casillero y una dirección sobre la cual
considerar, tal que, sean definidas como 'arriba', 'abajo', 'izquierda' y 'derecha' sobre el tablero"""
def pos_adyacente(pos, direccion):
    posicion = list(pos)
    if direccion == "arriba":
        posicion[0] = posicion[0] - 1 
    elif direccion == "abajo":
        posicion[0] = posicion[0] + 1
    elif direccion == "izquierda":
        posicion[1] = posicion[1] - 1
    elif direccion == "derecha":
        posicion[1] = posicion[1] + 1
    return (posicion[0], posicion[1])

"""Verifica, dada una posición y su dirección, si su adyacente es legal sobre los límites o rango del
tablero, que, como se conoce, se debe estar, computacionalmente, de 0 a 4 para filas y columnas"""
def es_limite_tablero(pos, direccion):
    if es_pos_legal(pos_adyacente(pos, direccion)) == True:
        return True
    else:
        return False

"""Verifica, dada una posición y su dirección, si, en principio, el casillero es comprendido por un
mosquetero, a lo cual, de ser cierto, se valida si el movimiento está en el límite del tablero y si
el mencionado tiene a un guardia, ya que, es la única jugada que puede efectuar el mosquetero, al no
poder, por regla, situarse en una posición ya vacía, con lo cual, se retorna como proceso válido"""
def es_mov_legal_mosquetero(pos, direccion):
    if ver_casillero(pos) != "M":
        raise ValueError("La pieza no es un mosquetero")
    else:
        if es_limite_tablero(pos, direccion):
            if ver_casillero(pos_adyacente(pos, direccion)) == "G":
                return True
        return False
    
"""Verifica, dada una posición y su dirección, si, en principio, el casillero es comprendido por un 
guardia, a lo cual, de ser cierto, se valida si el movimiento está en el límite del tablero y si el
mencionado tiene a un espacio vacío, ya que, es la única jugada que puede efectuar el guardia, al no
poder, por regla, desplazar una posición que tenga a un mosquetero, con lo que, se retorna True"""
def es_mov_legal_guardia(pos, direccion):
    if ver_casillero(pos) != "G":
        raise ValueError("La pieza no es un guardia")
    else:
        if es_limite_tablero(pos, direccion):
            if ver_casillero(pos_adyacente(pos, direccion)) == "-":
                return True
        return False

"""Verifica, dada una inspección del casillero, si la posición y dirección indicada es un movimiento
legal, ya sea, el casillero contenido por un mosquetero o por un guardia, y retorna el resultado"""
def es_mov_legal(pos, direccion):
    if ver_casillero(pos) == "M":
        return es_mov_legal_mosquetero(pos, direccion)
    elif ver_casillero(pos) == "G":
        return es_mov_legal_guardia(pos, direccion)

"""Verifica, dada una posición, si el jugador tiene al menos, un movimiento disponible, al ser que, 
según la iteración del diccionario de orientaciones, se retornará True si existe, en efecto, un 
movimiento que permita al jugador seguir vigente, verificando la legalidad y límite en el tablero"""
def es_mov_disponible(pos):
    for orientacion in orientaciones:
        if es_mov_legal(pos, orientacion) and es_limite_tablero(pos, orientacion):
            return True
    return False

"""Valida, ahora, dado un jugador, si tiene aún posibles movimientos a partir de la iteración de la
lista total de las posiciones del tablero, de ver_casillero y de es_mov_disponible según la posición"""
def tiene_mov_disponible(jugador):
    for posicion in lista_posiciones():
        if ver_casillero(posicion) == jugador and es_mov_disponible(posicion):
            return True
    return False

"""A partir de una posición en el tablero, se retornará las posibles orientaciones que se pueden dar,
si es que el movimiento es legal y si la posición adyacente también lo es, con lo cual, se añadirá
a la lista de movimientos y se retornará con las posibles orientaciones que se dispongan actualmente"""
def ver_posibles_movimientos(pos):
    movimientos = []
    for orientacion in orientaciones:
        if es_mov_legal(pos, orientacion) and es_pos_legal(pos_adyacente(pos, orientacion)):
            movimientos.append(orientacion)
    return movimientos

"""Devuelve, para la estrategia Primero el mejor, todos los posibles movimientos para el jugador como 
una lista de coordenadas (ubicación, dirección), siempre que las orientaciones sean posibles"""
def ver_movimientos_jugador(jugador):
    return [(i, j) for i in lista_posiciones() for j in ver_posibles_movimientos(i) if ver_casillero(i) == jugador and es_mov_disponible(i)]

"""Devuelve, para la estrategia aleatoria, todos los posibles movimientos para el jugador como una
lista, donde, se verificará que pertenezca al casillero del jugador y que la lista no sea vacía """
def ver_movimientos_jugador_aleatorio(jugador):
    movimientos = []
    for i in lista_posiciones():
        if ver_casillero(i) == jugador and ver_posibles_movimientos(i) != []:
            movimientos.append((i, ver_posibles_movimientos(i)))
    if movimientos != []:
        return movimientos

"""Se efectuará, dado el tablero global, el movimiento, tomando a la posición adyacente y reemplazando
su nuevo valor, marcando al anterior con '-', representando vacío, donde cabe mencionar que no se 
realiza una validación en este punto, ya que, la misma, sea efectúa en otra función personalizada"""
def efectuar_movimiento(pos, direccion):
    dir_ady = pos_adyacente(pos, direccion)
    tablero[dir_ady[0]][dir_ady[1]] = tablero[pos[0]][pos[1]]
    tablero[pos[0]][pos[1]] = "-"

"""Se describe el movimiento, dado un jugador, una posición y una dirección, tal que, tampoco se da
una validación en este punto, ya que, la misma, el movimiento dado se confirmará posteriormente"""
def describir_movimiento(jugador, pos, direccion):
    nueva_pos = pos_adyacente(pos, direccion)
    print(jugador, "se mueve", direccion, "de",\
          pos_a_str(pos), "a",\
          pos_a_str(nueva_pos) + ".\n")

"""Se define el módulo de la estrategia aleatoria, en donde, según los movimientos que se disponen,
que debe ser al menos uno, se tendrá que, se elegirá al azar alguna de lista y se devolverá como 
una tupla con el valor de la posición y de la orientación, siempre que sea legal en el tablero"""
def estrategia_aleatoria(jugador):
    import random
    if ver_movimientos_jugador_aleatorio(jugador) != None:
        elecciones = ver_movimientos_jugador_aleatorio(jugador)
        eleccion_aleatoria = random.choice(elecciones)
        return (eleccion_aleatoria[0], random.choice(eleccion_aleatoria[1]))

"""Se define el módulo de la estrategia Primero el mejor, en donde, según el jugador, se efectuará
el movimiento correspondiente, el mismo que será definido, según las condiciones, posteriormente"""
def estrategia_primero_el_mejor(jugador):
    if jugador == "M":
        return movimiento_mosquetero()
    if jugador == "G":
        return movimiento_guardia()

"""Se retorna a la jugada de la pc según el jugador actual y la dificultad, que en este caso, se 
asigna a la aleatoria como fácil y a Primero el mejor, como media, a través de la primera letra"""
def definir_jugada_pc(jugador, dificultad):
    if dificultad == "F":
        return estrategia_aleatoria(jugador)
    elif dificultad == "M":
        return estrategia_primero_el_mejor(jugador)

"""Se verifica si en el juego, han ganado los guardias, donde, a través de la iteración de todo el
tablero, se cuenta en filas y columnas a la cantidad de mosqueteros, donde, si se da en algún caso que
sea 3, habrán ganado, consecuentemente, los guardias; caso contrario, se retornará el valor False"""
def gana_guardia():
    for i in range(5):
        cont_fila = 0
        cont_columna = 0
        for j in range(5):
            if ver_casillero((i, j)) == "M":
                cont_fila = cont_fila + 1
            if ver_casillero((j, i)) == "M":
                cont_columna = cont_columna + 1
        if cont_fila == 3 or cont_columna == 3:
            return True
    return False

"""Se elegirá un jugador (mosquetero o guardia) para el usuario real en cierta partida que ejecute"""
def elegir_jugador():
    jugador = ""
    while jugador != "M" and jugador != "G":
        respuesta = input("\n¿Desea jugar como Mosquetero (M) o como Guardia (G)?: ").strip()
        if respuesta != "":
            jugador = respuesta.upper()[0]
    return jugador

"""Se definirá la dificultad (fácil o media) del juego, teniendo en cuenta, a las 2 estrategias"""
def elegir_dificultad():
    dificultad = ""
    while dificultad != "F" and dificultad != "M":
        respuesta = input("\nIndique dificultad facil (F) o media (M): ").strip()
        if respuesta != "":
            dificultad = respuesta.upper()[0]
    if dificultad == "F":
        nombre_estrategia = "\nDificultad fácil: Estrategia aleatoria"
    elif dificultad == "M":
        nombre_estrategia = "\nDificultad media: Estrategia Primero el mejor"
    tupla = dificultad, nombre_estrategia
    return tupla

"""Se obtendrá la validación del movimiento introducido por el jugador, tal que, deba cumplir 4
condiciones, que de ser ciertas, se verificará si el movimiento es legal o no, retornando en casos"""
def obtener_movimiento():
    direcciones = {"AR" : "arriba", "AB" : "abajo", "IZ" : "izquierda", "DE" : "derecha"}
    mov = input("Digite movimiento: ").upper().replace(' ', '')
    if (len(mov) >= 4
            and mov[0] in "ABCDE"
            and mov[1] in "12345"
            and mov[2:4] in "ARABIZDE"):
        pos = str_a_pos(mov[0:2])
        direccion = direcciones[mov[2:4]]
        if es_mov_legal(pos, direccion):
            return (pos, direccion)
    print("Movimiento ilegal: " + mov)
    return obtener_movimiento()

#--------------------------------------------------------------------------------------------------------------
#--------------------------------------------------------------------------------------------------------------
"""Se obtiene el valor de la heurística propuesta, que es que se verifica las posiciones actuales de
los mosqueteros y se evalúa el valor de los movimientos de acuerdo a la suma de distancias de estos, 
tal que, se devolverá un diccionario de key: tupla de coordenadas y orientación y value: heurística"""
def obtener_valor_jugada():
    movimientos = ver_movimientos_jugador("M")
    valor_movimientos = {}
    for posibles_mov in movimientos:
        pos_mosq = posiciones_mosquetero()
        pos_mosq.remove(posibles_mov[0])
        pos_mosq.append(pos_adyacente(posibles_mov[0], posibles_mov[1]))
        sum_x = abs(pos_mosq[0][0] - pos_mosq[1][0]) + abs(pos_mosq[1][0] - pos_mosq[2][0]) + abs(pos_mosq[0][0] - pos_mosq[2][0])
        sum_y = abs(pos_mosq[0][1] - pos_mosq[1][1]) + abs(pos_mosq[1][1] - pos_mosq[2][1]) + abs(pos_mosq[0][1] - pos_mosq[2][1])
        valor_mov = sum_x + sum_y
        valor_movimientos[posibles_mov] = valor_mov
    return valor_movimientos

"""Se devuelve un diccionario de cada posible movimiento asociado de los guardias asociado a la suma
de valores de los mosqueteros desde una jugada, teniendo en cuenta que se realiza este proceso desde
la copia del tablero, para no afectar al mismo en caso la cantidad de movimientos sea diferente de 0"""
def obtener_valor_enemigo():
    valores_enemigo = {}
    for mov_guardia in ver_movimientos_jugador("G"):
        copia_tab = get_tablero()
        efectuar_movimiento(mov_guardia[0], mov_guardia[1])
        movimientos_total = obtener_valor_jugada()
        if len(movimientos_total) != 0:
            suma = 0
            for m_move in movimientos_total:
                suma = suma + movimientos_total[m_move]
            suma = suma / len(movimientos_total)
            if suma in valores_enemigo.keys():
                valores_enemigo[suma] = valores_enemigo[suma] + [mov_guardia]
            elif suma not in valores_enemigo.keys():
                valores_enemigo[suma] = [mov_guardia]
        set_tablero(copia_tab)
    return valores_enemigo
#--------------------------------------------------------------------------------------------------------------
#--------------------------------------------------------------------------------------------------------------


"""Se define el movimiento del mosquetero, que seguirá la lógica de que, si se tiene una única jugada
disponible, se tomará y se retornará; caso contrario, se creará un diccionario de posibles jugadas con
su valor asociado, ordenados desde el más alto hasta el más bajo, retornando, por supuesto, el mayor"""
def movimiento_mosquetero():
    movimientos = ver_movimientos_jugador("M")
    if len(movimientos) == 1:
        return movimientos[0]
    valor_mov = dict(sorted(obtener_valor_jugada().items(), key = lambda kv: kv[1], reverse = True))
    for mov in valor_mov:
        return mov

"""Se define el movimiento del guardia, que seguirá la lógica de que, si se tiene una única jugada 
disponible, se tomará y se retornará; caso contrario, si no se han realizado movimientos hasta el 
momento, se elegirá uno al azar y se retornará; de no ser así, se deberá mover en la misma orientación
que las jugadas anteriores, ya que, se puede definir que, si es jugada aún se mantiene vigente, es que
ha tenido relativo éxito en el juego y por ende, es aconsejable seguir con dicho patrón seleccionado"""
def movimiento_guardia():
    movimientos = ver_movimientos_jugador("G")
    from random import randint
    if len(movimientos) == 1:
        return movimientos[0]
    valores_enemigo = dict(sorted(obtener_valor_enemigo().items(), key = lambda kv: kv[1], reverse = False))
    if len(valores_enemigo) == 0:
        mov_elegido = movimientos[randint(0,len(movimientos) - 1)]
        return mov_elegido
    for mov in valores_enemigo:
        mejor_pos = sorted(contador_pc.items(), key = lambda kv: kv[1], reverse = True)[0][0]
        for i in range(len(valores_enemigo[mov])):
                if valores_enemigo[mov][i][1] == mejor_pos:
                    return valores_enemigo[mov][i]
        return valores_enemigo[mov][randint(0, len(valores_enemigo[mov]) - 1)]

"""Dado un usuario y una dificultad, se realizará manualmente el movimiento o jugada determinada, tal
que, a continuación, de forma intercalada, se tenga el movimiento, si lo tiene, de algún guardia"""
def mover_mosquetero(usuario, dificultad):
    if usuario == "M":
        (pos, direccion) = obtener_movimiento()
        if ver_casillero(pos) == "M":
            if es_mov_legal(pos, direccion):
                efectuar_movimiento(pos, direccion)
                describir_movimiento("Mosquetero", pos, direccion)
        else:
            print("No se puede mover a esa posición")
            return mover_mosquetero(usuario, dificultad)
    else:
        (pos, direccion) = definir_jugada_pc("M", dificultad)
        efectuar_movimiento(pos, direccion)
        describir_movimiento("Mosquetero", pos, direccion)

"""Dado un usuario y una dificultad, se realizará manualmente el movimiento o jugada determinada, tal
que, a continuación, de forma intercalada, se tenga el movimiento, si lo tiene, de algún mosquetero"""
def mover_guardia(usuario, dificultad):
    if usuario == "G":
        (pos, direccion) = obtener_movimiento()
        if ver_casillero(pos) == "G":
            if es_mov_legal(pos, direccion):
                efectuar_movimiento(pos, direccion)
                describir_movimiento("Guardia", pos, direccion)
        else:
            print("No se puede mover a esa posición")
            return mover_guardia(usuario, dificultad)
    else:
        (pos, direccion) = definir_jugada_pc("G", dificultad)
        efectuar_movimiento(pos, direccion)
        agregar_movimiento(direccion)
        print(contador_pc)
        describir_movimiento("Guardia", pos, direccion)
        return tablero

"""Esta función acopla la llamada a la ejecución del programa, con lo cual, se elige al usuario,
dificultad, se inicializa el tablero y el contador de jugadas de la pc y, mientras existan movimientos
disponibles de ambas partes (mosqueteros y guardias), el juego se seguirá ejecutando y pidiendo que
se den movimientos, evaluando, por supuesto, si luego de alguno, se tiene algún ganador del juego"""
def ejecutar_programa():
    title01 = "\n¡BIENVENIDOS!"
    title02 = "\nLOS TRES MOSQUETEROS"
    print(title01)
    print(title02)
    print("-"*len(title02))
    usuario = elegir_jugador()
    dificultad = elegir_dificultad()
    tablero = crear_tablero()
    contador = crear_contador()
    
    print(dificultad[1]+"\n")
    imprimir_tablero()
    while True:
        if tiene_mov_disponible("M"):
            tablero = mover_mosquetero(usuario, dificultad[0])
            imprimir_tablero()
            if gana_guardia():
                print("Ganan los guardias")
                break
        else:
            print("Ganan los mosqueteros")
            break
        if tiene_mov_disponible("G"):
            tablero = mover_guardia(usuario, dificultad[0])
            imprimir_tablero()
        else:
            print("Ganan los mosqueteros")
            break
```
